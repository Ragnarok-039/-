```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from sqlalchemy import create_engine

db_config = {'user': 'challenge_user', # имя пользователя
             'pwd': 'aT00GAh944YU4Q1J4zbzryW6Yrz11iq54coz', # пароль
             'host': 'rc1a-eqy3nymy2wp6tbx2.mdb.yandexcloud.net', # адрес сервера
             'port': 6432, # порт подключения
             'db': 'data-analyst-challenge'}# название базы данных
 
# Формируем строку соединения с БД.
connection_string = 'postgresql://{}:{}@{}:{}/{}'.format(db_config['user'],
                                             db_config['pwd'],
                                             db_config['host'],
                                             db_config['port'],
                                             db_config['db'])
# Подключаемся к БД.
engine = create_engine(connection_string)

folder = os.getcwd()
sql_folder = fr'{folder}\SQL'

# Формируем sql-запрос(пример).
query = open(fr"{sql_folder}\yandex.sql").read()

# Выполняем запрос и сохраняем результат
# выполнения в DataFrame.
# Sqlalchemy автоматически установит названия колонок
# такими же, как у таблицы в БД.
data_raw = pd.io.sql.read_sql(query, con = engine)

print(data_raw.head())
```

Ниже мы видим, что уроки менее 5 секунд присутствуют в нашей выборке. Графики показывают количество таких уроков в разрезе по месяцам и дате.

```python
plt.rcParams["figure.figsize"] = (30,10)

data_raw['lesson_month'] = data_raw['lesson_datetime'].dt.month
data_raw['lesson_date'] = data_raw['lesson_datetime'].dt.date

df_count_by_month = data_raw[['delta_seconds', 'lesson_month']].groupby('lesson_month')['lesson_month'].agg('count')
df_count_by_date = data_raw[['delta_seconds', 'lesson_date']].groupby('lesson_date')['lesson_date'].agg('count')

print(df_count_by_month.head())
print(df_count_by_date.head())

df_count_by_month.plot(kind = 'bar')
plt.title('Общее количество уроков со слишком быстрым (менее 5 секунд) временем прохождения по месяцам', fontsize=20)
plt.ticklabel_format(style = 'plain', axis='y')
plt.xlabel('Месяц прохождения урока', horizontalalignment='center', fontsize=16)
plt.show()

df_count_by_date.plot(kind = 'bar')
plt.title('Общее количество уроков со слишком быстрым (менее 5 секунд) временем прохождения по дате', fontsize=20)
plt.ticklabel_format(style = 'plain', axis='y')
plt.xlabel('Дата прохождения урока', horizontalalignment='center', fontsize=16)
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (30,10)

df_count_by_lesson = data_raw[['delta_seconds', 'lesson_id']].groupby('lesson_id')['lesson_id'].agg('count')
df_5 = df_count_by_lesson.sort_values(ascending=False)[:10]

print(df_count_by_lesson.head())

df_count_by_lesson.plot(kind = 'bar')
plt.title('Общее количество уроков со слишком быстрым (менее 5 секунд) временем прохождения по имени урока', fontsize=20)
plt.ticklabel_format(style = 'plain', axis='y')
plt.xlabel('Имя урока', horizontalalignment='center', fontsize=16)
plt.show()

df_5.plot(kind = 'bar', width=0.75)
plt.title('Топ 10 по количеству уроков со слишком быстрым (менее 5 секунд) временем прохождения по имени урока', fontsize=20)
plt.ticklabel_format(style = 'plain', axis='y')
plt.xlabel('Имя урока', horizontalalignment='center', fontsize=16)
plt.show()
```
